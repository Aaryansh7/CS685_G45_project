```python
# Imports
import numpy as np
import pandas as pd
from tqdm import tqdm  
import requests
import pandas as pd
```

```python
dframe = pd.DataFrame()
for year in tqdm(range(1957,2022)):
    for month in range(13):
        url = 'https://www.n2yo.com/browse/?y=' + str(year) + '&m=' + str(month)
        html = requests.get(url).content
        df_list = pd.read_html(html)
        df = df_list[2]
        if df.empty == True:
            continue

        dframe = dframe.append(df, ignore_index=True)

dframe.drop_duplicates(subset ="Name", keep = False, inplace = True)
dframe.to_csv('/content/drive/MyDrive/CS685_project/data/launch_dates.csv')
dframe.head()
```
