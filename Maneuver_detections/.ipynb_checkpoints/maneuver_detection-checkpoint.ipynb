```python
df = pd.read_csv('data(1)/yaogan9B_delta_data.csv') 
df.drop([df.columns[0]], axis=1)
```

```python
normalized_df=(df-df.min())/(df.max()-df.min())
normalized_df.drop([normalized_df.columns[0]], axis=1, inplace=True)
normalized_df.head()
```

```python
d_sma_list = list(normalized_df.iloc[:]['d_sma'])
d_ecc_list = list(normalized_df.iloc[:]['d_ecc'])
d_inc_list = list(normalized_df.iloc[:]['d_inc'])
d_argument_list = list(normalized_df.iloc[:]['d_argument_perigee'])
d_raan_list = list(normalized_df.iloc[:]['d_raan'])
d_mean_list = list(normalized_df.iloc[:]['d_mean_anomaly'])
d_r_x_list = list(normalized_df.iloc[:]['d_r_x'])
d_r_y_list = list(normalized_df.iloc[:]['d_r_y'])
d_r_z_list = list(normalized_df.iloc[:]['d_r_z'])
d_pos_list = list(normalized_df.iloc[:]['d_norm_r'])
d_v_x_list = list(normalized_df.iloc[:]['d_v_x'])
d_v_y_list = list(normalized_df.iloc[:]['d_v_y'])
d_v_z_list = list(normalized_df.iloc[:]['d_v_z'])
d_vel_list = list(normalized_df.iloc[:]['d_norm_v'])

time_series = np.arange(len(d_sma_list))


dsma_denoise = list(denoise_wavelet(np.array(d_sma_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
decc_denoise = list(denoise_wavelet(np.array(d_ecc_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dinc_denoise = list(denoise_wavelet(np.array(d_inc_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
darg_denoise = list(denoise_wavelet(np.array(d_argument_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
draan_denoise = list(denoise_wavelet(np.array(d_raan_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dmean_denoise = list(denoise_wavelet(np.array(d_mean_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
drx_denoise = list(denoise_wavelet(np.array(d_r_x_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dry_denoise = list(denoise_wavelet(np.array(d_r_y_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
drz_denoise = list(denoise_wavelet(np.array(d_r_z_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dpos_denoise = list(denoise_wavelet(np.array(d_pos_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dvx_denoise = list(denoise_wavelet(np.array(d_v_x_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dvy_denoise = list(denoise_wavelet(np.array(d_v_y_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dvz_denoise = list(denoise_wavelet(np.array(d_v_z_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
dvel_denoise = list(denoise_wavelet(np.array(d_vel_list), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))

#MAKE CSV FROM ABOVE DATASHEET

dict = {
        'dsma_denoise':dsma_denoise, 
        'decc_denoise':decc_denoise, 
        'dinc_denoise': dinc_denoise,
        'darg_denoise': darg_denoise,
        'draan_denoise': draan_denoise,
        'dmean_denoise':dmean_denoise,
        'drx_denoise': drx_denoise, 
        'dry_denoise': dry_denoise, 
        'drz_denoise': drz_denoise,
        'dpos_denoise': dpos_denoise, 
        'dvx_denoise': dvx_denoise, 
        'dvy_denoise': dvy_denoise, 
        'dvz_denoise':dvz_denoise,
        'dvel_denoise':dvel_denoise,
       }
                  

dframe = pd.DataFrame(dict)   
#dframe.to_csv('/content/drive/MyDrive/CS685_project/data/yaogan9A.csv')
dframe.head()
```

```python
X = dsma_denoise
Y = list(np.zeros(len(X)))

X = np.array(list(zip(X, Y)))
```

```python
range_n_clusters = [2, 3]
```

```python
for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

```python
df = pd.read_csv('data(1)/yaogan9B.csv')
df.head()
```

```python
gravitation_param = 3.986004418e14
rev_day2rad_sec = 7.27220522e-5
def sma_compute(n):
    n_ = n*(2*np.pi)/86400
    denominator = n_**(2/3)
    numerator = gravitation_param**(1/3)
    return (numerator/denominator)/1000
semi_major_axis = []
for i in tqdm(range(df.shape[0])):
    semi_major_axis.append(sma_compute(df['MEAN MOTION(n)'][i]))
df['sma'] = semi_major_axis
df.head()
```

```python
semi_denoise = list(denoise_wavelet(np.array(semi_major_axis), method='BayesShrink', mode='soft', wavelet_levels=3, wavelet='sym8', rescale_sigma='True'))
print(semi_denoise)
```

```python
time_series = np.arange(200)
plt.plot(time_series, semi_denoise[:200], label='semi-major axis variation')
plt.xlabel('TLE')
plt.ylabel('semi-major axis(km)')
plt.title('sma variation curve')
plt.show()
```

```python
# compute distance matrix for kmeans 

distance_matrix = np.full((200, 200), -1, dtype=float)
for i in range(distance_matrix.shape[0]):
    for j in range(distance_matrix.shape[1]):
        distance_matrix[i][j] = np.sqrt( (X[i][0] - X[j][0])**2 +  (X[i][1] - X[j][1])**2 + (X[i][2] - X[j][2])**2 + (X[i][3] - X[j][3])**2 )
```

```python
matfig = plt.figure(figsize=(15,10))
plt.matshow(distance_matrix, cmap='viridis',fignum=matfig.number)
plt.colorbar()
plt.show()
```
